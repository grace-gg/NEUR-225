```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
url1 = 'https://raw.githubusercontent.com/grace-gg/neur-225/main/acc_fa_hw5.csv'
acc_fa = np.loadtxt(url1, delimiter = ',')
url2 = 'https://raw.githubusercontent.com/grace-gg/neur-225/main/acc_hits_hw5.csv'
acc_hits = np.loadtxt(url2, delimiter = ',')
```

```python
#prints the shape of each array
print(acc_fa.shape)
print(acc_hits.shape)
```

```python
#variable called time that has the same length as acc_hits and acc_fa, with equally spaced values between "zero" and "four" (the length of each trial, in seconds).
time = len(acc_hits), len(acc_fa)
time = np.linspace(0,4,8001)
print(time)
```

```python
#create an expression to determine the sampling rate of the data using the knowledge that you have 8001 samples, and the total length of each sample is 4 seconds.
8001/4
```

```python
#plot the first trial of acc_hits in one subplot, and the first trial of acc_fa in another subplot. Plot your time variable on the x-axis. Label your axes.
fig, ax = plt.subplots(1,2,figsize=(15,5))

ax[0].plot(time,acc_hits[0])  
ax[0].set_xlabel('Time [s]')   
ax[0].set_ylabel('Voltage [\mu V]')   
ax[0].set_title ('First Trial of Hits')       
ax[1].plot(time,acc_fa[0]) 
ax[1].set_xlabel('Time[s]')  
ax[1].set_ylabel('Voltage [\mu V]') 
ax[1].set_title('First Trial of False Alarm')       
```

In a new Markdown Cell, describe what you think the dominant oscillation (in Hz) in the LFP might be. Does this differ between the first trial of acc_hits and the first trial of acc_fa?

```python
# Calculate and plot the ERP for acc_hits
plt.plot(time, acc_hits.mean(0))     
plt.xlabel('Time [s]')          
plt.ylabel('Voltage [\mu V]')
plt.title('ERP of Hits')  
plt.vlines(2, -50, 50, color = 'black')
plt.hlines(0, time[0], time[-1], color = 'black')
```

Interpret your ERP for hits. Do you see any evidence of consistent activity in the LFP surrounding screen touch (vertical black bar)?

After the vertical black bar, there is an increasingly amount of activity.There are much larger oscillations that occur after this 2 second mark compared to before. This means that two seconds into the sample, neurons started firing as the screen was touched.
