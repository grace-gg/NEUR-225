```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
#make a predictors_conditioning variable that contains shock onset and tone onset. 
#Reference our in-class notebook from 4/12/23 to do this - specifically, the Code Cell underneath the section that says Modeling our Fear Data. 
#These are the same data that we used for this notebook, so you can use this exact code to re-create your shock_array and tone_array variables. 
#Reference how we made the predictors variable for model1 and model3 in that notebook 
#you should have the same predictors variable as model3 for your predictors_conditioning variable here (i.e., the two variables should be identical).
shock_array = np.zeros(10458)
for i in range(10458):
  if i >= (shock_index[0]-15)*15 and i <= (shock_index[0]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[1]-15)*15 and i <= (shock_index[1]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[2]-15)*15 and i <= (shock_index[2]+30)*15:
    shock_array[i] = 1
  else:
    shock_array[i] = 0

tone_array = np.zeros(10458)
for i in range(10458):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array[i] = 1
  else:
    tone_array[i] = 0

from pandas import DataFrame as df
import statsmodels.api as sm
from statsmodels.genmod.families import Poisson
from statsmodels.genmod.families.links import identity, log
from scipy.stats import chi2
from statsmodels.distributions.empirical_distribution import ECDF

first_trace = conditioning_traces[:,0]

predictors_conditioning = df(data={'Intercept': np.ones_like(tone_array), 'tone onset': tone_array})

predictors_conditioning['shock onset'] = shock_array
model3 = sm.GLM(first_trace, predictors_conditioning, family=Poisson())
model3_results = model3.fit() 
b1 = model3_results.params
print('b1:')
print(b1)

     

     
```

```python
#We want to now run our model for each ROI in our conditioning_traces array, correlate the predicted trace vs. the actual trace, 
#and store the correlation coefficient in a new variable called r_model_conditioning
#You will need to first create your r_model_conditioning variable using np.zeros 
#remember that this time, we have more than 10 ROIs, so you will need to change the number inside of the parentheses. 
#You will also need to change the number of times your loop runs from for i in range(10) to loop through all of the ROIs in your conditioning_traces array. 
#In a new Code Cell, create your r_model_conditioning variable, and run your for loop. 
#Each loop should extract an ROI from conditioning_traces, run sm.GLM between the trace and the predictors variable you created, fit the model, 
#create a predicted trace, get a correlation coefficient using pearsonr, and store the correlation coefficient in a new column of r_index_conditioning.

predictors_conditioning['shock onset'] = shock_array
r_model_conditioning = np.zeros(63)

for i in range(63):
  trace_temp = conditioning_traces[:,i]
  model = sm.GLM(trace_temp, predictors_conditioning, family=Poisson())
  model_results = model.fit()
  predicted_trace = model_results.predict()
  r = pearsonr(trace_temp, predicted_trace)
  r_model_conditioning[i] = r.statistic
```

```python
#run another loop that does the same thing for your extinction_traces array. 
#This time, you should make a variable called r_index_extinction to store your correlation coefficients. 
#You should use your extinction_predictors variable as the predictor for your model (this contains values for tone onset during the extinction session).
shock_array = np.zeros(14209)
for i in range(14209):
  if i >= (shock_index[0]-15)*15 and i <= (shock_index[0]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[1]-15)*15 and i <= (shock_index[1]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[2]-15)*15 and i <= (shock_index[2]+30)*15:
    shock_array[i] = 1
  else:
    shock_array[i] = 0

tone_array = np.zeros(14209)
for i in range(14209):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array[i] = 1
  else:
    tone_array[i] = 0

first_trace = extinction_traces[:,0]

extinction_predictors = df(data={'Intercept': np.ones_like(shock_array), 'shock onset': shock_array})
model3 = sm.GLM(first_trace, extinction_predictors, family=Poisson())
model3_results = model3.fit() 
b1 = model3_results.params
print('b1:')
print(b1)

extinction_predictors['tone onset'] = tone_array
r_model_extinction = np.zeros(63)

for i in range(63):
  trace_temp = extinction_traces[:,i]
  model = sm.GLM(trace_temp, extinction_predictors, family=Poisson())
  model_results = model.fit()
  predicted_trace = model_results.predict()
  r = pearsonr(trace_temp, predicted_trace)
  r_model_extinction[i] = r.statistic
```

```python
#use the pearsonr function to correlate your r_index_conditioning and r_index_extinction variabl
pearsonr(r_model_conditioning, r_model_extinction)
```

```python
#make a scatter plot with r_index_conditioning on the x-axis, and r_index_extinction on the y-axis. Label your axes.
plt.scatter(r_model_conditioning,r_model_extinction)
plt.xlabel('r_model_conditioning')
plt.ylabel('r_model_extinction')
```
