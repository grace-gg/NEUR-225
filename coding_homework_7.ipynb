```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
url1 = 'https://raw.githubusercontent.com/grace-gg/neur-265/main/spikes_left_trial1.csv'
spikes_left_trial1 = np.loadtxt(url1, delimiter = ',')

url2 = 'https://raw.githubusercontent.com/grace-gg/neur-265/main/spikes_right_trial1.csv'
spikes_right_trial1 = np.loadtxt(url2, delimiter = ',')

url3 = 'https://raw.githubusercontent.com/grace-gg/neur-265/main/spikes_left_trial2.csv'
spikes_left_trial2 = np.loadtxt(url3, delimiter = ',')

url4 = 'https://raw.githubusercontent.com/grace-gg/neur-265/main/spikes_right_trial2.csv'
spikes_right_trial2 = np.loadtxt(url4, delimiter = ',')
```

```python
print(len(spikes_left_trial1))
print(len(spikes_right_trial1))
print(len(spikes_left_trial2))
print(len(spikes_right_trial2))
```

```python
#write an expression to determine the firing rate of this neuron during each trial 
#(first and second left turn trials, first and second right turn trials).
rate1 = len(spikes_left_trial1)/30
print(rate1)
rate2 = len(spikes_left_trial2)/30
print(rate2)
rate3 = len(spikes_right_trial1)/30
print(rate3)
rate4 = len(spikes_right_trial2)/30
print(rate4)
```

Do you think this neuron fires more before left turn trials, or right turn trials?

I think this neuron fires at a similar rate in the left and right turn trials. The main difference I see between the firing rates is that the first left turn trial has a higher firing rate than the second left turn trial and the second right turn trial has a higher firing rate than the first right turn trial.

```python
#make two subplots, and plot the spike train data from trial 1 in your first subplot, and the spike train data from trial 2 in your second subplot. 
#You should have dots that correspond to action potentials in two rows (one for left turn, one for right turn) in each subplot. Label your axes.

fig, ax = plt.subplots(1,2,figsize=(12,5))

ax[0].plot(spikes_left_trial1, np.ones_like(spikes_left_trial1), '.')
ax[0].plot(spikes_right_trial1, 2 * np.ones_like(spikes_right_trial1), '.')
ax[0].title.set_text('Trial 1')
ax[0].legend({'Left', 'Right'})
ax[0].set_xlabel('Time [s]')
ax[0].set_yticks([])

ax[1].plot(spikes_left_trial2, np.ones_like(spikes_left_trial2), '.')
ax[1].plot(spikes_right_trial2, 2 * np.ones_like(spikes_right_trial2), '.')
ax[1].title.set_text('Trial 2')
ax[1].legend({'Left', 'Right'})
ax[1].set_xlabel('Time [s]')   
ax[1].set_yticks([])
            
```

Interpret the graph from step 8. Does this neuron seem more "bursty", or more regular-firing, during left or right turn trials?

This neuron seems more bursty during both left and right turn trials. In each trial, there are bursts of firing and then large gaps of nothing for less than five seconds. Following that gap, there is more bursting.

```python
#calculate inter-spike interval (ISI) arrays for each of your four spike arrays using the np.diff() function. 
#Make sure to create a variable name for each ISI array.

ISIsleft1 = np.diff(spikes_left_trial1)
ISIsleft2 = np.diff(spikes_left_trial2)
ISIsright1 = np.diff(spikes_right_trial1)
ISIsright2 = np.diff(spikes_right_trial2)
```

```python
#make four histograms - one for each of your four ISI arrays. 
#Set the bins for your histograms to start at time 0, and end at time 2 s, with a bin size equal to 0.01 s. Label your axes.

bins = np.arange(0, .2, 1e-2)  

plt.hist(ISIsleft1, bins)       
plt.xlim([0, 0.25])           
plt.xlabel('ISI [s]')             
plt.title('Left Turn Trial 1')      
```

```python
bins = np.arange(0, .2, 1e-2)  

plt.hist(ISIsleft2, bins)
plt.xlim([0, 0.25])
plt.xlabel('ISI [s]')
plt.title('Left Turn Trial 2')
```

```python
bins = np.arange(0, .2, 1e-2)  

plt.hist(ISIsright1, bins)
plt.xlim([0, 0.25])
plt.xlabel('ISI [s]')
plt.title('Right Turn Trial 1')
```

```python
bins = np.arange(0, .2, 1e-2)  

plt.hist(ISIsright2, bins)
plt.xlim([0, 0.25])
plt.xlabel('ISI [s]')
plt.title('Right Turn Trial 2')
```

Interpret your histograms. Does this neuron tend to fire in "bursts"? Does it have higher "burstiness" during left or right turn trials?

This neuron tends to fire in bursts between 0 and 0.05 seconds in all left and right turn trials. However, the neuron seems to have more bustiness during the left turn trials (with exception to the outlier burst in trial 2 of the right turn around 0.05 seconds). Additionally, all histograms have tails indicating that there is less to no bursting after around 0.05 seconds.

```python
#calculate the firing rate across time for each of your four arrays (trial 1 left-turn, trial 1 right-turn, trial 2 left-turn, and trial 2 right-turn). 
#Set your time bins equal to start at 0 s, end at 30 s (the length of the waiting period), with a bin size equal to 1 s. 
#Create two subplots, and plot the firing rate over time for each trial in each of the two subplots. 
#For example, your first subplot should have the firing rate across time for trial 1 left-turn, and trial 1 right-turn, 
#and your second subplot should have the firing rate across time for trial 2 left-turn, and trial 2 right-turn. 
#Make the line for right-turn trials blue, and the line for left-turn trials red. Label your axes.

rate1 = len(spikes_left_trial1)/30
print(rate1)
rate2 = len(spikes_left_trial2)/30
print(rate2)
rate3 = len(spikes_right_trial1)/30
print(rate3)
rate4 = len(spikes_right_trial2)/30
print(rate4)

fig, ax = plt.subplots(1,2,figsize=(12,5))

binsize = 1                                                                   
fr_bins = np.arange(0,30,binsize)                                             
fr_left1 = np.zeros(len(fr_bins)-1)
fr_left2 = np.zeros(len(fr_bins)-1)    
fr_right1 = np.zeros(len(fr_bins)-1)
fr_right2 = np.zeros(len(fr_bins)-1)                                                                          
for i in range(len(fr_bins)-1):                                               
  spk_index1 = (spikes_left_trial1 > fr_bins[i]) & (spikes_left_trial1 < fr_bins[i+1])
  spk_index2 = (spikes_left_trial2 > fr_bins[i]) & (spikes_left_trial2 < fr_bins[i+1])
  spk_index3 = (spikes_right_trial1 > fr_bins[i]) & (spikes_right_trial1 < fr_bins[i+1]) 
  spk_index4 = (spikes_right_trial2 > fr_bins[i]) & (spikes_right_trial2 < fr_bins[i+1])        
  true_values1 = spikes_left_trial1[spk_index1] 
  true_values2 = spikes_left_trial2[spk_index2] 
  true_values3 = spikes_right_trial1[spk_index3]  
  true_values4 = spikes_right_trial2[spk_index4]                                             
  fr_left1[i] = (len(true_values1))/binsize            
  fr_left2[i] = (len(true_values2))/binsize  
  fr_right1[i] = (len(true_values3))/binsize
  fr_right2[i] = (len(true_values4))/binsize

  ax[0].plot(fr_bins[:-1], fr_left1, color = 'red')
  ax[0].plot(fr_bins[:-1], fr_right1, color = 'blue')
  ax[0].set_xlabel('Time [s]')
  ax[0].set_ylabel('Firing Rate [Hz]')
  ax[0].set_title('Trial 1')
  ax[0].legend({'Left', 'Right'})

  ax[1].plot(fr_bins[:-1], fr_left2, color = 'red')
  ax[1].plot(fr_bins[:-1], fr_right2, color = 'blue')
  ax[1].set_xlabel('Time [s]')
  ax[1].set_ylabel('Firing Rate [Hz]')
  ax[1].set_title('Trial 2')
  ax[1].legend({'Left', 'Right'})
```
